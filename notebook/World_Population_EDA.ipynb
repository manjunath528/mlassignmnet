Here, We have seen that some numeric attributes are showing as the object so we convert them into the numeric datatype

```python
df['Urban  Pop %'] = pd.to_numeric(df['Urban  Pop %'].str.rstrip('%'), errors='coerce')
df['Urban Population'] = pd.to_numeric(df['Urban Population'], errors='coerce')
df["Country's Share of  World Pop"] = pd.to_numeric(df["Country's Share of  World Pop"].str.rstrip('%'), errors='coerce')
df['Yearly %   Change'] = pd.to_numeric(df['Yearly %   Change'].str.rstrip('%'),errors='coerce')
```

```python
data = data.drop(columns=['yearly_change','urban_population','country_share_of_world','world_population'],axis=1)
```

```python
numeric_features = [feature for feature in data.columns if data[feature].dtype != 'O']
```

```python
corr_matrix = data[numeric_features]
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
numeric_columns = data.select_dtypes(include=['int64','float64'])
Q1 = numeric_columns.quantile(0.25)
Q3 = numeric_columns.quantile(0.75)
threshold_multiplier = 1.5
IQR = Q3-Q1
lower_bound = Q1 - threshold_multiplier * IQR
upper_bound = Q3 + threshold_multiplier * IQR
outliers = ((numeric_columns < lower_bound) | (numeric_columns > upper_bound)).any(axis=1)
```

```python
clean_dat = data[~outliers]
```

```python
clean_dat.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
clean_dat['country'] = LabelEncoder().fit_transform(clean_dat['country'])
```
